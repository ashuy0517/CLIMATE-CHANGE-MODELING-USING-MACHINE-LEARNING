```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
data = pd.read_csv("climate_nasa.csv")
data.head()
```

```python
print("Shape:", data.shape)
data.info()
data.describe()
data.isnull().sum()
```

```python
# Fill missing values for 'commentsCount' with its mean
data['commentsCount'] = data['commentsCount'].fillna(data['commentsCount'].mean())

# Fill missing values for 'text' with an empty string
data['text'] = data['text'].fillna('')
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(data[['likesCount', 'commentsCount']].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Between Climate Variables")
plt.show()
```

```python
data['date'] = pd.to_datetime(data['date'])
data.set_index("date")['likesCount'].plot(figsize=(10,4))
plt.title("Climate Trend Over Time")
plt.show()
```

```python
data.hist(figsize=(12,6), bins=30)
plt.show()
```

```python
target = "likesCount"  # Using 'likesCount' as the target column

X = data.drop(target, axis=1)
y = data[target]
```

```python
X_numeric = X.select_dtypes(include=np.number) # Select only numeric columns
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_numeric)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

```python
model = RandomForestRegressor(random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2  = r2_score(y_test, y_pred)

print(" Model Evaluation")
print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", np.sqrt(mse))
print("R Score:", r2)
```

```python
plt.figure(figsize=(6,5))
plt.scatter(y_test, y_pred, alpha=0.6)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs Predicted Climate Values")
plt.show()
```

```python
importance = pd.Series(model.feature_importances_, index=X_numeric.columns)

plt.figure(figsize=(10,5))
importance.sort_values().plot(kind='barh')
plt.title("Feature Importance")
plt.show()
```

```python
future = pd.DataFrame({'commentsCount': [10, 20, 30]}) # Create a dummy future DataFrame

future_scaled = scaler.transform(future)
future_pred = model.predict(future_scaled)

print("Future Predictions:", future_pred)
```
